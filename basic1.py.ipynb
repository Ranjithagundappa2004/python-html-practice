```python
class fan:
    def __init__(self):
        self.brand="Usha"
        self.color="white"
        self.cost=1500
        self.wings=3

    def on(self):
        print("fan is on")

    def rotate(self):
        print("fan is rotating")

    def off(self):
        print("fan is off")

f1=fan()
print(f1.brand)
print(f1.color)
print(f1.cost)
print(f1.wings)

f1.on()
f1.rotate()
f1.off()
```

```python
class car:
    def __init__(self):
        self.brand="TATA"
        self.color="white"

    def start(self):
        print("car is starting")
        print(self.color)
        print(self)

c1=car()
print(c1.brand)
print(c1.color)
c1.start()
print(c1)
```

```python
class cat:
    def __init__(self):
        self.colour="white"
        self.legs=4
        self.ears=2

    def sound(self):
        print("cat is sounding")

c1=cat()
print(c1.colour)
print(c1.legs)
print(c1.ears)
c1.sound()
```

```python
class flower:
    def __init__(self):
        self.name="rose"
        self.petals=20
        self.colour="red"

    def bloom(self):
        print("flower is blooming")
        print(self.petals)
        print(self)

f1=flower()
print(f1.petals)
print(f1.name)
print(f1.colour)
f1.bloom()
print(f1)
```
